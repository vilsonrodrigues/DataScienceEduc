Lendo CSV

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import datetime as dt
from pandas import Timestamp
```

```
table=pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/arquivosCSV/lopUseTurmasSubLista.csv")
table.head()
```

Apagando colunas

```
table.drop(columns={"Unnamed: 0","titulo","descricaoTurma","nome"},inplace=True)
table.head(2)
```

```
table.dtypes
```

```
table.count()
```

```
table["data"] = pd.to_datetime(table["data"])
table["idTurma"].dropna(axis=0, how='any',inplace=True)
table.dropna(inplace=True)
table = table.drop_duplicates()
table.reset_index()
table.sort_values(by=["idAluno","questao","data"])
#table.tail()
table.count()
```

```
table["data"] = pd.to_datetime(table["data"])
table.reset_index()
```

```
table.count()

newdf = pd.DataFrame()

prev=table.iloc[0]

for index, row in table.iterrows():
	if index == 0 or pd.isnull(row['data']) or pd.isnull(prev['data']): 
		continue # skip first row and rows with null data
	a=prev["data"]
	b=row["data"]
	if prev["idAluno"] != row["idAluno"]:
		newdf = newdf.append(prev)
	elif (b-a).days > 0:
		newdf = newdf.append(prev)
	elif (b-a).total_seconds() >= 3:
		newdf = newdf.append(prev)
	prev = row

#o newdf que contém o df filtrado	
print(newdf.count())
```

```
newdf.to_csv(r"lopSubmissoes.csv")
```
